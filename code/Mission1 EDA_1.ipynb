```python
# 이미지 리스트 정렬
import glob

path = "./dataset/"

L10 = sorted(glob.glob(path+"/L2_10" + '/*'))
L12 = sorted(glob.glob(path+"/L2_12"+'/*'))
L15 = sorted(glob.glob(path+"/L2_15"+'/*'))
L20 = sorted(glob.glob(path+"/L2_20"+'/*'))
L21 = sorted(glob.glob(path+"/L2_21"+'/*'))
L24 = sorted(glob.glob(path+"/L2_24"+'/*'))
L25 = sorted(glob.glob(path+"/L2_25"+'/*'))
L27 = sorted(glob.glob(path+"/L2_27"+'/*'))
L3 =  sorted(glob.glob(path+"/L2_3"+'/*'))
L30 = sorted(glob.glob(path+"/L2_30"+'/*'))
L33 = sorted(glob.glob(path+"/L2_33"+'/*'))
L34 = sorted(glob.glob(path+"/L2_34"+'/*'))
L39 = sorted(glob.glob(path+"/L2_39"+'/*'))
L40 = sorted(glob.glob(path+"/L2_40"+'/*'))
L41 = sorted(glob.glob(path+"/L2_41"+'/*'))
L44 = sorted(glob.glob(path+"/L2_44"+'/*'))
L45 = sorted(glob.glob(path+"/L2_45"+'/*'))
L46 = sorted(glob.glob(path+"/L2_46"+'/*'))
L50 = sorted(glob.glob(path+"/L2_50"+'/*'))
L52 = sorted(glob.glob(path+"/L2_52"+'/*'))
```

```python
import os

alllabel = os.listdir(path)
illustlabel=['L2_10','L2_15','L2_20','L2_21','L2_25','L2_27','L2_30','L2_33','L2_34','L2_39','L2_40','L2_44','L2_45','L2_46','L2_52']
pictlabel=['L2_3','L2_12','L2_24','L2_41','L2_50']
```

```python
# 이미지 개수 시각화

import matplotlib.pyplot as plt
import numpy as np

value = []

for i in alllabel:
  value.append(len(os.listdir(path+i)))

plt.figure(figsize=(15, 8))
bar = plt.bar(alllabel, value, color = "skyblue")

#숫자 보이게
for rect in bar:
  height = rect.get_height()
  plt.text(rect.get_x() + rect.get_width()/2.0, height, '%d' % height, ha='center', va='bottom', size = 12)

plt.show()
```

### 2. 이미지 사이즈 및 차원

```python
all=[L10,L15,L20,L21,L25,L27,L30,L33,L34,L39,L40,L44,L45,L46,L52,L3,L12,L24,L41,L50]
illust=[L10,L15,L20,L21,L25,L27,L30,L33,L34,L39,L40,L44,L45,L46,L52]
pict=[L3,L12,L24,L41,L50]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import pandas as pd
plt.figure(figsize=(60, 32))

for j in range(len(illust)):
   imgsize=[]
   plt.subplot(4,5,j+1)
   for i in tqdm(illust[j]):
      k=plt.imread(i)
      imgsize.append(k.shape[:2])
   imgsize=pd.Series(imgsize)
   x=np.arange(len(imgsize.value_counts()))
   values=imgsize.value_counts().values
   years=imgsize.value_counts().index
   plt.bar(x, values,color = "blue")
   plt.xticks(x, years)
   plt.title(f'{illustlabel[j]}')

for l in range(5):
   imgsize=[]
   plt.subplot(4,5,j+2+l)
   for t in tqdm(pict[l]):
      p=plt.imread(t)
      imgsize.append(p.shape[:2])
   imgsize=pd.Series(imgsize)
   x=np.arange(len(imgsize.value_counts()))
   values=imgsize.value_counts().values
   years=imgsize.value_counts().index
   plt.bar(x, values,color = "red")
   plt.xticks(x, years)
   plt.title(f'{pictlabel[l]}')

plt.show()
```
