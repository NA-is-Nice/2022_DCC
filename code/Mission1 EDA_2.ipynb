```python
from PIL import Image,ImageFilter
path=['L2_10', 'L2_12', 'L2_15', 'L2_20', 'L2_21', 'L2_24', 'L2_25', 'L2_27', 'L2_3', 'L2_30', 'L2_33', 'L2_34', 'L2_39', 'L2_40', 'L2_41', 'L2_44', 'L2_45', 'L2_46', 'L2_50', 'L2_52']
```

```python
# 원본 dataset 안에 있는 파일들 확인
import os
import pandas as pd
for dirpath,dirnames,filenams in os.walk('./dataset'):
    print(dirpath,dirnames)
```

```python
import glob
import natsort
import matplotlib.pyplot as plt
imgpath=[]
for i in range(1,20):
    img_paths=glob.glob(f'./dataset/{path[i]}/*.jpg')
    img_paths=natsort.natsorted(img_paths)
    imgpath.append(img_paths)
```

```python
# 폴더별 사진 살펴보기
for n in range(0,19):
    print(f"label : {path[n]}")
    plt.figure(figsize=(12,30))
    
    for m in range(0,80):
        plt.subplot(56, 20, m+1) 
        original=Image.open(imgpath[n][m])
        plt.imshow(original)
        plt.axis("off")
    plt.show()
```
