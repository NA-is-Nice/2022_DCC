```python
#  
import os
import numpy as np
from PIL import Image

#    
targerdir = r"new_dataset"  #   

files = os.listdir(targerdir)

format = [".jpg", ".png", ".jpeg", "bmp", ".JPG", ".PNG", "JPEG", "BMP"]  #    
for (path, dirs, files) in os.walk(targerdir):
    for file in files:
        if file.endswith(tuple(format)):
            image = Image.open(path + "/" + file)
            print(image.filename)
            print(image.size)

            if image.mode != 'RGB':
                image = image.convert('RGB')
                print("RGB    !")

            image = image.resize((100,100))
            image.save("new_dataset/" + file)
            # image.save(file)

            print(image.size)

        else:
            print(path)
            print("InValid", file)

#    
image_path = 'new_dataset/100/'

img_list = os.listdir(image_path)  #     
img_list_jpg = [img for img in img_list if img.endswith(".jpg")]  #   

print("img_list_jpg: {}".format(img_list_jpg))

img_list_np = []

for i in img_list_jpg:
    img = Image.open(image_path + i)
    img_array = np.array(img)
    img_list_np.append(img_array)
    print(i, "   - :", img_array.shape)  #     (XX)
    print(img_array.T.shape) # (XX)

img_np = np.array(img_list_np)  #  numpy 
np.save('./new_dataset', img_np)  # x_save.npy
print(img_np.shape)
print(":    ")
```
